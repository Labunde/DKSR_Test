```python
raw = pd.read_json('NextBikeCologne.json')
raw2 = raw.values.tolist()
raw3 = pd.DataFrame(raw2)[0].apply(lambda x: json.loads(x))
export = pd.DataFrame.from_dict(raw3[0])
```

```python
#Zeilenweise Extrahierung
for ll in range(1,len(raw3)):
    new_line = pd.DataFrame.from_dict(raw3[ll])
    export = pd.concat([export,new_line])
```
